```python
import control as co

# 개방루프 전달함수 정의
num = [100] # 분자 계수
den = [1, 5, 6] # 분모 계수: s^2 + 5s + 6
G = co.TransferFunction(num, den)

# 피드백 전달함수 정의 (H(s) = 1)
H = co.TransferFunction([1], [1])

# 폐루프 전달함수 계산
T = co.feedback(G, H)

print("Closed-loop transfer function: ", T)
```

```python
import matplotlib.pyplot as plt
import numpy as np


# 시간 벡터 생성
t = np.linspace(0, 10, 1000)

# unit step 입력에 대한 응답 계산
t, y = co.step_response(T, t)

# 그래프 그리기
plt.figure()
plt.plot(t, y)
plt.title('Step response for the closed-loop transfer function')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Bode plot 생성
plt.figure()
co.bode_plot(T, dB=True)
plt.show()
```
